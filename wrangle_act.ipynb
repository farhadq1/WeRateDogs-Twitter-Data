## Define

- Find enteries where name is wrong or missing and replace by correct name where possible

## Code

```python
#There are names starting with lowercases e.g 'a', 'quite', etc these which shows incorrect names
# Creating a mask for enteries where name is lower and containes the word named in text
lower_having_named = twarch_clean.name.str.islower() & twarch_clean.text.str.contains('named')
```

```python
for i in twarch_clean[lower_having_named].index:
    temp = twarch_clean[lower_having_named].loc[i].text
    temp=pd.DataFrame([[temp]])
    twarch_clean.loc[i,'name'] = temp[0].str.extract('named\s(\w+)',expand=True)[0][0]
```

```python
#Similarly creating mask for lowecase names with name following the phrase 'name is' in tweet text
lower_having_nameis = twarch_clean.name.str.islower() & twarch_clean.text.str.contains('name is')
```

```python
for j in twarch_clean[lower_having_nameis].index:
    temp = twarch_clean[lower_having_nameis].loc[j].text
    temp=pd.DataFrame([[temp]])
    twarch_clean.loc[j,'name'] = temp[0].str.extract('name is\s(\w+)',expand=True)[0][0]
```

```python
## Replacing the remaining lowercase names with None since names for them could not be extracted
name_is_lower = twarch_clean.name.str.islower() 
for k in twarch_clean[name_is_lower].index:
    twarch_clean.loc[k,'name'] = "None"
```

```python
## Replacing all None with nan
twarch_clean.name=twarch_clean.name.replace('None',np.nan)
```
